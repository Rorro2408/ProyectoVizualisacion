# Pre procesamiento de datasets.

```python
import numpy as np
import pandas as pd
```

```python
df_años = pd.read_csv("/content/drive/MyDrive/tcc_ceds_music.csv")
df_spotify = pd.read_csv("/content/drive/MyDrive/dataset_spotify.csv")
```

```python
df_años.head()
```

```python
df_spotify.head()
```

```python
df_spotify.columns
```

```python
columnsToDrop = ["Unnamed: 0", "track_id", "explicit", "key", "loudness", "mode", "speechiness", "acousticness",
                 "instrumentalness", "liveness", "valence", "tempo", "time_signature"]
```

```python
df_spotify = df_spotify.drop(columnsToDrop, axis=1)
```

```python
df_spotify
```

```python
df_spotify["artists"] = df_spotify["artists"].str.lower()
df_spotify["track_name"] = df_spotify["track_name"].str.lower()
df_spotify
```

```python
df_spotify = df_spotify.rename(columns={'artists': 'artist_name'})
df_spotify
```

```python
df_años.columns
```

```python
columnsToDrop = ["Unnamed: 0", "lyrics", "len", "dating", "violence", "world/life", "night/time", "shake the audience", "family/gospel",
                 "romantic", "communication", "obscene", "music", "movement/places", "light/visual perceptions", "family/spiritual",
                 "like/girls", "sadness", "loudness", "acousticness", "instrumentalness", "valence", "topic", "age"]
```

```python
df_años = df_años.drop(columnsToDrop, axis=1)
```

```python
df_años
```

```python
df_complete_song_data = pd.merge(df_spotify, df_años, on=['artist_name', 'track_name'], how='inner')
df_complete_song_data
```

```python
columnsToDrop = ["track_genre" ,"danceability_y", "energy_y"]
df_complete_song_data = df_complete_song_data.drop(columnsToDrop, axis=1)
df_complete_song_data
```

```python
df_complete_song_data = df_complete_song_data.rename(columns={"duration_ms": "duration", "danceability_x":"danceability", "energy_x": "energy"})
df_complete_song_data
```

```python
df_noDuplicates = df_complete_song_data.drop_duplicates(subset=['artist_name', 'track_name'])
df_noDuplicates
```

```python
unique_genres = df_noDuplicates['genre'].unique()
unique_genres
```

```python
df_noDuplicates.to_csv('/content/drive/MyDrive/Complete_song_data_no_duplicates.csv', index=False)
```
